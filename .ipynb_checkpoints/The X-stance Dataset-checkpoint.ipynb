# The X-stance Dataset

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import collections
```

## Train Data

```python
train_data = pd.read_json('xstance-master/data/xstance-data-v1.0/train.jsonl',lines=True,
                        orient='columns')
train_data.head(5)
```

```python
print('total number of records', len(train_data))
train_label = collections.Counter(train_data['label'])
print('Stance bifurcation', train_label)
lang_label = collections.Counter(train_data['language'])
print('Language bifurcation', lang_label)

train_frq = train_data.groupby(['language','label']).size().reset_index()
train_frq=train_frq.rename(columns={0:'Count'})

train_frq= pd.pivot_table(data=train_frq , index=['language'], columns=['label'],values=['Count'])
train_frq.reset_index(inplace=True)
train_frq.columns =['language','AGAINST','FAVOR']
train_frq
```

## Validation Data

```python
val_data = pd.read_json('xstance-master/data/xstance-data-v1.0/valid.jsonl', lines=True, orient ='columns')
val_data.head(5)
```

```python
print('Total Number of records', len(val_data))
val_labels = collections.Counter(val_data['label'])
print('Stance bifurcation', val_labels)
val_lang_label = collections.Counter(val_data['language'])
print('Language bifurcation', val_lang_label)

#val_frq = val_data.groupby(["language", "label"]).size()
val_frq = val_data.groupby(['language', 'label']).size().reset_index()
val_frq=val_frq.rename(columns={0:'Count'})
val_frq = pd.pivot_table(data=val_frq, index=['language'], columns=['label'], values=['Count'])
val_frq.reset_index(inplace=True)
val_frq.columns = ['language', 'AGAINST','FAVOR']
val_frq
```

```python
val_frq = val_data.groupby(['language', 'label']).size().reset_index()
val_frq=val_frq.rename(columns={0:'Count'})
val_frq
```

## Test Data

```python
test_data = pd.read_json('xstance-master/data/xstance-data-v1.0/test.jsonl',lines=True, orient='columns')
test_data.head(5)
```

```python
print('total number of records', len(test_data))
test_labels = collections.Counter(test_data['label'])
print('Stance bifurcation',test_labels)
test_lang_label = collections.Counter(test_data['language'])
print('Language bifurcation',test_lang_label)

test_frq = test_data.groupby(['language','label']).size().reset_index()
test_frq = test_frq.rename(columns={0:'Count'})

test_frq = pd.pivot_table(data=test_frq, index=['language'], columns=['label'], values=['Count'])
test_frq.reset_index(inplace=True)
test_frq.columns=['language','AGAINST','FAVOR']
test_frq
```

```python
fig, axes = plt.subplots(1,3, figsize=(20,5))

train_frq.plot(x='language', y=['AGAINST','FAVOR'], kind='bar', ax=axes[0], title='Training Set')
val_frq.plot(x='language', y=['AGAINST','FAVOR'], kind='bar', ax=axes[1], title='Validation Set')
test_frq.plot(x="language", y=["AGAINST", "FAVOR"], kind="bar", ax=axes[2], title='Test Set')
```
