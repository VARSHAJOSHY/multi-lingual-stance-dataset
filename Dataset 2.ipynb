```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import collections
```

```python
france_data = pd.read_csv('MultilingualStanceDetection-master\dataset\lepen_fr.csv')
```

```python
france_data.head(5)
```

```python
fr1_label_counts = collections.Counter(france_data['Stance'])
print("Test Data", fr1_label_counts)
fr1_label_counts = collections.Counter(france_data['Set'])
print("Test Data", fr1_label_counts)
```

```python
france_data2 = pd.read_csv('MultilingualStanceDetection-master\dataset\macron_fr.csv')
france_data2.head(5)
```

```python
fr2_label_counts = collections.Counter(france_data2['Set'])
print("Test Data", fr2_label_counts)
fr2_label_counts = collections.Counter(france_data2['Stance'])
print("Test Data", fr2_label_counts)
```

```python
italian_data = pd.read_csv('MultilingualStanceDetection-master/dataset/referendum_it.csv')
italian_data.head(5)
```

```python
ita_label_counts = collections.Counter(italian_data['Set'])
print("Test Data", ita_label_counts)
ita_label_counts = collections.Counter(italian_data['Stance'])
print("Test Data", ita_label_counts)
```

```python
france_frq = france_data.groupby(['Stance']).size().reset_index()
france_frq=france_frq.rename(columns={0:'Count'})
france_frq['language'] ='French'

france_frq = pd.pivot_table(data=france_frq, index=['language'], columns=['Stance'], values=['Count'])
france_frq.reset_index(inplace=True)
france_frq.columns = ['language', 'agains','favor','none']

print(france_frq)

print("\n")
france2_frq1 = france_data2.groupby(['Stance']).size().reset_index()
france2_frq1=france2_frq1.rename(columns={0:'Count'})
france2_frq1['language'] ='French'

france2_frq1 = pd.pivot_table(data=france2_frq1, index=['language'], columns=['Stance'], values=['Count'])
france2_frq1.reset_index(inplace=True)
france2_frq1.columns = ['language', 'agains','favor','none']

print(france2_frq1)

print("\n")
ital_frq1 = italian_data.groupby(['Stance']).size().reset_index()
ital_frq1=ital_frq1.rename(columns={0:'Count'})
ital_frq1['language'] ='Italian'

ital_frq1 = pd.pivot_table(data=ital_frq1, index=['language'], columns=['Stance'], values=['Count'])
ital_frq1.reset_index(inplace=True)
ital_frq1.columns = ['language', 'agains','favor','none']

print(ital_frq1)
```

```python
frames=[france_frq,france2_frq1,ital_frq1]
db_final = pd.concat(frames)
db_final = db_final.groupby(['language']).sum().reset_index()
db_final
```

```python
db_final.plot(x='language', y=['agains','favor','none'], kind='bar',  title='Data Set')
```
